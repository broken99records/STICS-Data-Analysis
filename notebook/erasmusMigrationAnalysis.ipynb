```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv(
    "../data/erasmus_student_migration.csv",
    sep=",",
    on_bad_lines="skip"
)

df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.columns
```

```python
country_counts = df[' Participant Nationality'].value_counts()
country_counts


country_counts.plot(kind='bar', figsize=(10,5))

plt.title('Frequency of participant Nationality Countries')
plt.xlabel('Participant Nationality')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
' Receiving Country Code'
receiving_country = df[' Receiving Country Code'].value_counts()
receiving_country


receiving_country.plot(kind='bar', figsize=(10,5))

plt.title('Most Popular Receipient Countries')
plt.xlabel('Receipient Nation')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
field_of_education = df[' Field of Education                                                      '].value_counts().head(10)

# Plot bar chart
field_of_education.plot(kind='bar', figsize=(10,5))

plt.title('Top 10 Fields of Education')
plt.xlabel('Field of Education')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
