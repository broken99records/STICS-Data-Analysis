```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/erasmus_student_migration.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.columns
```

```python
country_counts = df[' destination_country'].value_counts()
country_counts


country_counts.plot(kind='bar', figsize=(10,5))

plt.title('Frequency of Destination Countries')
plt.xlabel('Destination Country')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
